```python
import cryptocompare
import matplotlib.pyplot as plt
from datetime import datetime
from matplotlib.animation import FuncAnimation
from time import sleep
```

```python
def getPrice(name):
    return cryptocompare.get_price(name,currency='USD')[name]['USD']
```

```python
#to calculate percent difference
def getPriceDiff(name):
    inVal = getPrice(name)
    sleep(1)
    finVal = getPrice(name)
    diff = (finVal-inVal)/inVal
    return diff*100 
```

```python
print(getPriceDiff('ETH'))
```

```python
plt.style.use('ggplot')
%matplotlib notebook
```

```python
timeVal = []
price = []
def animate(i):
  timeVal.append(datetime.now())
  price.append(getPriceDiff('ETH'))
  plt.title("Ethereum")
  plt.xlabel("Time")
  plt.ylabel("Price")
  plt.plot_date(timeVal,price,linestyle ="solid",ms=0)
  plt.tight_layout()
```

```python
ani = FuncAnimation(plt.gcf(),animate,interval=1000)
plt.show()
```
